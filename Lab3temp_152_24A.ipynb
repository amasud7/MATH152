```python
# Enter your Python code here
x = sp.symbols('x')
f = sp.sqrt(x)
g = (x-3)**2
plots = plot(f, (x, 0, 5), line_color = 'r', show=False)
plots.extend(plot(g, (x, 0, 5), line_color='b', show=False))
plots.show()
```

#### 1b

```python
# Enter your Python code here
h = f - g
c = 2*sp.pi.evalf()*(x-1)
bounds = sp.solve(h, x)
upper_bound = bounds[0]
lower_bound = bounds[1]
integrand = h * c
integral = sp.integrate(integrand, (x, (lower_bound, upper_bound)))
print(f"The volume rotated around x = 1 is: {integral.evalf()}")
```

#### 1c

```python
# Enter your Python code here
R = 4 - g
r = 4 - f
integrand = sp.pi.evalf()*(R**2) - sp.pi.evalf()*(r**2)
integral = sp.integrate(integrand, (x, (lower_bound, upper_bound)))
print(f"The volume rotated around y = 4 is: {integral.evalf()}")
```
