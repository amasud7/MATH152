```python
r = sp.symbols('r')
x = sp.symbols('x')
function = (sp.pi) * ((r**2)-(x**2) - 1)
upperbound = sp.sqrt((r**2)-1)
lowerbound = -1 * (sp.sqrt((r**2)-1))
s = sp.integrate(function, (x, (lowerbound, upperbound)))
s
```

```python
ffunction = 2 * (x**2)
gfunction = abs(x**2 - 9)
plots = plot(ffunction, (x, -5, 5), line_color='r', show=False)
plots.extend(plot(gfunction, (x, -5, 5), line_color='b', show=False))
plots.show()
```

#### 2b

```python
# Finding points of intersection (set equations set to each other)
func1 = (x**2)-(2*x**2)-9 # function using the positive of abs function
func2 = (-x**2) - (2*x**2)+9 # function using the negative of abs function
intersection = sp.solve(func2, x)
print(f"The x points of intersection are x = {intersection[0].evalf()} and x = {intersection[1].evalf()}")
```

#### 2c

```python
# Area enclosed by the 2 functions
upperbound = intersection[1]
lowerbound = intersection[0]
upper_function = -x**2 + 9
lower_function = 2*x**2
subtraction = upper_function - lower_function
integral = sp.integrate(subtraction, (x, lowerbound, upperbound))
print(f"The approximate area of the region enclosed with the two functions is {integral.evalf()}")
```
