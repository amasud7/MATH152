```python
import numpy as np
import sympy as sp
from sympy.plotting import (plot, plot_parametric)
```

### Question 1

#### 1a

```python
# Given the normal distribution, plot.
x = sp.symbols('x')
f = (1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x/3)**2))
plots = plot(f, (x, -9, 9))
```

#### 1b Left Endpoint Approximation

```python
b = 6
a = -6
n = 500
deltaX = (b-a)/n
# arange(start, end (exclusive), step size)
# linspace(start, end (exclusive), # of points)

# part 1:
x1 = np.arange(a, b, deltaX)

# part 2:
y = (1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x1/3)**2))

# part 3
sum_yL = sum(y) * deltaX
print(f"Left endpoint is: {sum_yL.evalf()}")
```

### Question 2 Right Endpoint Approximation

```python
x1 = np.arange(a+deltaX, b+deltaX, deltaX)
y = (1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x1/3)**2))
sum_yR = sum(y) * deltaX
print(f"Right endpoint is: {sum_yR.evalf()}")
```

### Question 3 Midpoint Approximation

#### 3a

```python
x1 = np.arange(a+(deltaX/2), b-(deltaX/2)+deltaX, deltaX)
y = (1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x1/3)**2))
sum_yM = sum(y) * deltaX
print(f"Midpoint is: {sum_yM.evalf()}")
```

#### 3b

```python
avg = (sum_yL+sum_yR)/2
print(f"The average of left and right is not equal to the midpoint")
print(f"left and right: {avg.evalf()}")
print(f"midpoint: {sum_yM.evalf()}")
```

### Question 4 Trapezoid Approximation

#### 4a

```python
from scipy.integrate import trapz
x1 = np.arange(a, b+deltaX, deltaX)
f = ((1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x1/3)**2)))
trap = trapz(f, x1)
print(trap.evalf())
```

#### 4b

```python
print(f"The trapezoid approximation is not equal to the average of left and right")
print(f"left and right: {avg.evalf()}")
print(f"trapz: {trap.evalf()}")
```

### Question 5 Simpson's Rule

```python
from scipy.integrate import simps
x1 = np.arange(a, b+deltaX, deltaX)
f = ((1/(3*sp.sqrt(2*sp.pi))) * (sp.E**((-1/2)*(x1/3)**2)))
s = simps(f, x1)
print(f"Simps: {s.evalf()}")
```

### Question 6 Errors

```python
actual = 0.954499736103642

left = abs(actual - sum_yL.evalf())
print(f"Error for left: {left}")

right = abs(actual - sum_yR.evalf())
print(f"Error for right: {right}")

mid = abs(actual - sum_yM.evalf())
print(f"Error for mid point: {mid}")

trap = abs(actual - trap.evalf())
print(f"Error for trapezoid: {trap}")

simps = abs(actual - s.evalf())
print(f"Error for simps: {simps}")
```
